#1. Introdução


A otimização é um dos pilares fundamentais do aprendizado de máquina, pois determina a qualidade do treinamento dos modelos. No contexto de redes neurais, a otimização é essencial para ajustar os pesos das conexões entre os neurônios de forma que a rede aprenda padrões e generalize bem para novos dados.

A função objetivo em problemas de aprendizado supervisionado geralmente é uma função de perda (ou custo) que mede o quão bem o modelo está se saindo ao prever as saídas corretas a partir das entradas fornecidas. A otimização tem como finalidade minimizar essa função de perda, ajustando iterativamente os pesos da rede neural para melhorar seu desempenho.

##Importância da otimização no aprendizado de máquina
Os algoritmos de aprendizado de máquina dependem de técnicas de otimização para encontrar os melhores parâmetros do modelo. Sem um método eficiente para otimização, o modelo pode:

* Convergir lentamente: Pode levar muitas iterações para encontrar bons parâmetros.

* Ficar preso em mínimos locais: Pode parar em uma solução subótima em vez do melhor conjunto de parâmetros.

* Não generalizar bem: Pode aprender padrões específicos dos dados de treinamento sem conseguir extrapolar corretamente para novos dados.

##O papel do gradiente descendente no ajuste de pesos de modelos de redes neurais
O gradiente descendente é um dos algoritmos de otimização mais utilizados para minimizar a função de perda em redes neurais. Ele trabalha iterativamente ajustando os pesos da rede na direção oposta ao gradiente da função de custo, buscando minimizar o erro.

Em cada iteração, o algoritmo realiza os seguintes passos:

* Calcula o gradiente da função de perda em relação aos pesos: Isso indica a taxa de mudança da função de perda quando os pesos são ajustados.

* Atualiza os pesos na direção contrária ao gradiente: Como o gradiente aponta na direção de maior crescimento da função de perda, tomar o passo na direção contrária reduz essa função.

* Itera esse processo até que a função de perda atinja um mínimo aceitável: Dependendo da taxa de aprendizado e da condição de parada, o algoritmo pode continuar ajustando os pesos por várias épocas.

Matematicamente, a atualização dos pesos $w$ em uma rede neural é dada por:
$w := w - \eta \nabla J(w)$

Onde:

 $\eta$ é a taxa de aprendizado, que controla o tamanho dos passos dados na direção do gradiente.

$\nabla J(w)$ é o gradiente da função de custo em relação aos pesos $w$ .

Dessa forma, o gradiente descendente possibilita que a rede neural aprenda padrões e se ajuste aos dados, tornando-se um método essencial para o treinamento de modelos de aprendizado profundo.


#2. Tipos de Gradiente Descente
O gradiente descendente pode ser aplicado de diferentes formas, dependendo de como os dados de treinamento são utilizados para atualizar os pesos da rede neural. Existem três principais abordagens, que são as seguintes:

## 2.1. Gradiente Descendente em Batch

O gradiente descendente em batch, também chamado de gradiente descendente em lote, utiliza todos os exemplos do conjunto de treinamento para calcular o gradiente antes de atualizar os pesos. A atualização ocorre uma vez por época.

**Vantagens**:

- Atualização mais estável, pois o gradiente é calculado considerando todos os dados.

- Convergência suave, reduzindo oscilações indesejadas nos pesos.

**Desvantagens**:

- Alto custo computacional, pois o gradiente é calculado usando todo o conjunto de dados.

- Pode ser ineficaz para conjuntos de dados muito grandes.

## 2.2. Gradiente Descendente Estocástico (SGD)

No gradiente descendente estocástico (SGD), os pesos são atualizados após cada amostra individualmente. Isso significa que a atualização ocorre de forma mais frequente.

**Vantagens**:

- Mais eficiente para grandes conjuntos de dados.

- Pode escapar de mínimos locais devido à variabilidade das atualizações.

**Desvantagens**:

- Oscilação na convergência, tornando o aprendizado menos estável.

- Pode precisar de uma taxa de aprendizado menor para estabilizar a aprendizagem.

## 2.3. Gradiente Descendente Mini-Batch

O gradiente descendente mini-batch é um compromisso entre os dois anteriores. Ele divide o conjunto de treinamento em pequenos lotes (mini-batches) e atualiza os pesos com base no gradiente médio de cada lote.

**Vantagens**:

- Mais eficiente que o batch, pois reduz a carga computacional.

- Reduz oscilações excessivas do SGD, melhorando a estabilidade da aprendizagem.

**Desvantagens**:

- Pode introduzir alguma variabilidade nas atualizações, dependendo do tamanho do mini-batch.

## 2.4. Diferenças Fundamentais Entre os Métodos

**Batch:** Atualização mais estável, mas lenta.

**SGD**: Rápido, mas pode oscilar.

**Mini-Batch**: Balanceia eficiência e estabilidade.



# Gradiente Descendente em Batch

##Definição e Funcionamento

O gradiente descendente em batch, também conhecido como gradiente descendente em lote, é um método onde a atualização dos pesos ocorre após calcular o gradiente da função de perda considerando todos os exemplos do conjunto de treinamento. Isso significa que, em cada época, o modelo processa todos os dados antes de atualizar os pesos. Nocedal e Wright (2006) descrevem que essa abordagem reduz as oscilações do processo de otimização, garantindo um caminho mais estável para a convergência, sendo amplamente utilizada em cenários onde a estabilidade do aprendizado é mais importante do que a velocidade de atualização dos pesos.

##Fórmulas Matemáticas

A atualização dos pesos é dada por: $ w := w - \eta \nabla J(w) $

Onde:

- $ w $ representa o vetor de pesos da rede.
- $ \eta $ é a taxa de aprendizado.
- $ \nabla J(w) $ é o gradiente da função de custo calculado considerando todos os dados de treinamento.

A derivada parcial da função de custo em relação a cada peso  $ w_j $ é calculada como:

$$ \frac{\partial J(w)}{\partial w_j} = \frac{1}{N} \sum_{i=1}^{N} \nabla J(w; x_i, y_i) $$

Onde:

- $ N $ é o número total de exemplos no conjunto de treinamento.
- $(x_i, y_i)$ representam os exemplos individuais.


## Onde o Gradiente Descendente em Batch é Utilizado?

1. **Redes Neurais Profundas com Pequenos Conjuntos de Dados**  
   - Em redes neurais onde o tamanho do conjunto de dados é gerenciável em termos computacionais, o gradiente descendente em batch pode ser usado para garantir uma convergência mais estável e previsível.  
   - Exemplos incluem modelos acadêmicos ou aplicações médicas onde há um volume limitado de dados de pacientes.

2. **Problemas de Regressão e Classificação com Pequenos Conjuntos de Dados**  
   - O gradiente descendente em batch pode ser usado para ajustar modelos como regressão linear e logística quando o conjunto de dados não é massivo.  
   - Aplicações incluem previsão de preços imobiliários, diagnóstico médico supervisionado e análise financeira com bases de dados limitadas.

3. **Treinamento de Modelos de Aprendizado Profundo em GPUs Potentes**  
   - Em algumas situações, quando há grande poder computacional disponível, como em servidores otimizados para deep learning, o batch pode ser viável para modelos mais complexos, pois maximiza o aproveitamento de processamento em GPUs.

4. **Otimização de Modelos em Sistemas de Controle e Engenharia**  
   - Aplicado em problemas de otimização, como controle de robôs ou otimização de processos industriais, onde a estabilidade e precisão da convergência são desejáveis.

## Características Dessa Técnica

- **Sensibilidade à Taxa de Aprendizado**  
  - Como o batch processa todos os dados antes de atualizar os pesos, escolher uma taxa de aprendizado muito grande pode resultar em saltos grandes demais, dificultando a convergência. Já uma taxa muito pequena pode tornar o treinamento lento.
  
- **Influência de Dados Não Balanceados**  
  - Se o conjunto de dados tiver muitas classes desbalanceadas, o cálculo do gradiente pode ser enviesado em direção às classes mais representadas, tornando necessário o uso de técnicas como reamostragem ou balanceamento dos dados.

- **Convergência Mais Suave**  
  - Comparado ao gradiente descendente estocástico (SGD), o método em batch reduz variações abruptas nas atualizações dos pesos, resultando em trajetórias de aprendizado mais consistentes.




# 4. Gradiente Descendente Estocástico (SGD)  



## Definição e Funcionamento  
O gradiente descendente estocástico (SGD) é uma variação do algoritmo de gradiente descendente onde os pesos são atualizados após o cálculo do gradiente de uma única amostra de treinamento em vez de todo o conjunto de dados. Isso significa que, para cada amostra individualmente, ocorre uma atualização imediata dos pesos da rede. Bottou (2010) explica que essa abordagem pode ser mais eficiente em conjuntos de dados grandes, pois permite atualizações frequentes dos parâmetros sem a necessidade de processar o dataset inteiro, tornando o treinamento mais rápido e acessível para modelos de aprendizado de máquina em larga escala.

Dessa forma, o SGD introduz um grau de aleatoriedade na otimização, pois cada atualização pode seguir uma direção diferente dependendo da amostra processada. Embora essa variação cause oscilações no aprendizado, ela pode ajudar o modelo a escapar de mínimos locais e encontrar um ótimo global mais rapidamente. Hastie et al. (2009) destacam que essa característica do SGD pode ser útil especialmente em problemas não convexos, onde a busca pelo ótimo global pode ser prejudicada por mínimos locais indesejados.

## Fórmulas Matemáticas
A atualização dos pesos no SGD é dada por: $w := w - \eta \nabla J(w; x_i, y_i)$

Onde:
- $ w $ representa o vetor de pesos da rede.
- $ \eta $ é a taxa de aprendizado.
- $ \nabla J(w; x_i, y_i) $ é o gradiente da função de custo calculado **para uma única amostra** $ (x_i, y_i) $ do conjunto de dados.

A derivada parcial da função de custo em relação a cada peso $ w_j $ é:

$$
\frac{\partial J(w)}{\partial w_j} = \nabla J(w; x_i, y_i)
$$

onde $ (x_i, y_i) $ é uma amostra aleatória extraída do conjunto de treinamento.


## Onde o Gradiente Descendente Estocástico é Utilizado?

1. **Grandes Conjuntos de Dados**  
   - Quando o conjunto de dados é muito grande para ser processado de uma vez, o SGD permite que o modelo seja treinado de forma eficiente, processando um exemplo por vez.  
   - Exemplos incluem treinamento de modelos em deep learning sobre grandes bases de dados como **ImageNet** ou **bases de texto massivas** para NLP.

2. **Problemas de Aprendizado Online**  
   - Em aplicações onde os dados chegam em fluxo contínuo (streaming de dados), o SGD pode ser usado para atualizar os pesos em tempo real.  
   - Exemplos: **sistemas de recomendação** e **previsão de séries temporais em tempo real**.

3. **Treinamento de Redes Neurais Profundas**  
   - Devido à sua eficiência computacional, o SGD é amplamente utilizado no treinamento de **redes neurais profundas**, principalmente quando combinado com técnicas como **momentum** e **taxas de aprendizado adaptativas**.


## Características Dessa Técnica
- **Alto Grau de Oscilação**: Como as atualizações são feitas após cada amostra, há uma grande variabilidade na direção do aprendizado. Para reduzir esse problema, técnicas como **momentum** e **decaimento da taxa de aprendizado** podem ser usadas.
  
- **Possibilidade de Escapar de Mínimos Locais**: Diferente do batch, que pode ficar preso em mínimos locais, o SGD pode ajudar o modelo a sair desses pontos devido à sua natureza estocástica.

- **Necessidade de Mais Épocas**: Como cada iteração usa apenas uma amostra, normalmente são necessárias mais épocas para o modelo convergir em comparação ao batch.



#5. Gradiente Descendente Mini-Batch





## Definição e Funcionamento
O gradiente descendente mini-batch é uma abordagem intermediária entre o batch e o SGD. Nesse método, o conjunto de treinamento é dividido em pequenos lotes chamados mini-batches, e os pesos da rede são atualizados após o cálculo do gradiente desses lotes. Segundo Ruder (2016), essa abordagem equilibra a estabilidade da otimização do método batch com a eficiência computacional do SGD, tornando-se uma escolha popular em aplicações de aprendizado profundo. Além disso, os métodos de otimização modernos, como o Adam, se beneficiam do mini-batch ao combinar a adaptabilidade do SGD com a estabilidade de um gradiente suavizado.

Esse método busca um equilíbrio entre a eficiência computacional e a estabilidade do treinamento, reduzindo as oscilações do SGD e melhorando o tempo de convergência do batch. Em aplicações práticas, Sutskever et al. (2013) demonstraram que o mini-batch pode acelerar significativamente o aprendizado ao permitir que os modelos aproveitem melhor a computação paralela em GPUs. O uso de Batch Normalization em redes neurais profundas pode ser combinado com mini-batches para estabilizar o treinamento e reduzir a covariância interna dos gradientes. Assim, o mini-batch se destaca como uma alternativa viável para problemas de grande escala, balanceando custo computacional e qualidade da otimização.

## Fórmulas Matemáticas
A atualização dos pesos no mini-batch é dada por: $w := w - \eta \nabla J(w; B)$

Onde:
- $w $ representa o vetor de pesos da rede.
- $ \eta $ é a taxa de aprendizado.
- $ B $ é um mini-lote contendo $ m $ amostras do conjunto de treinamento.
- $ \nabla J(w; B) $ é o gradiente médio calculado sobre o mini-batch $ B $.

A derivada parcial da função de custo em relação a cada peso $ w_j $ é:

$$
\frac{\partial J(w)}{\partial w_j} = \frac{1}{m} \sum_{i=1}^{m} \nabla J(w; x_i, y_i)
$$

onde $ m $ é o número de amostras dentro do mini-batch.

## Onde o Gradiente Descendente Mini-Batch é Utilizado?

1. **Treinamento de Modelos de Deep Learning**  
   - O mini-batch é amplamente utilizado no treinamento de **redes neurais profundas** porque melhora a eficiência computacional e reduz a variabilidade do aprendizado.  
   - Exemplos incluem modelos **CNNs para visão computacional** e **transformers para NLP**.

2. **Modelos de Aprendizado Supervisionado em Larga Escala**  
   - Em problemas como classificação e regressão em conjuntos de dados grandes, o mini-batch permite um equilíbrio entre precisão e eficiência.

3. **Treinamento em Ambientes de Computação Paralela**  
   - Ao processar mini-batches em GPUs e TPUs, é possível otimizar o desempenho do treinamento. Muitos frameworks modernos como **TensorFlow** e **PyTorch** são otimizados para esse tipo de processamento.



## Outras Informações Relevantes  
- **Escolha do Tamanho do Mini-Batch**:  
  - Um tamanho pequeno pode tornar o treinamento mais ruidoso, semelhante ao SGD.  
  - Um tamanho grande se aproxima do batch, perdendo a vantagem de processamento eficiente.  
  - Valores comuns: 32, 64, 128 (depende da capacidade de hardware disponível).

- **Menor Custo Computacional que o Batch**:  
  - Como os cálculos são feitos com subconjuntos de dados, o uso de memória e tempo de processamento é mais eficiente.

- **Redução da Oscilação em Relação ao SGD**:  
  - A média dos gradientes no mini-batch suaviza a atualização dos pesos, reduzindo as oscilações abruptas que ocorrem no SGD.



#6. Comparação Prática dos Métodos


O código a seguir foi desenvolvido para comparar de forma prática os métodos de otimização Batch, Mini-Batch e SGD. Para garantir reprodutibilidade, uma semente fixa foi definida, permitindo que os experimentos possam ser replicados com os mesmos resultados.

Utilizamos o dataset Fashion-MNIST, que contém imagens de roupas e acessórios, passando por um processo de normalização para melhorar a estabilidade do treinamento. Criamos três DataLoaders, cada um representando um método de otimização: Batch, que processa todo o conjunto antes de atualizar os pesos; SGD, que atualiza os pesos após cada amostra; e Mini-Batch, que realiza atualizações em pequenos lotes de amostras.

A rede neural utilizada é uma MLP simples, contendo duas camadas ocultas (128 e 64 neurônios) ativadas por ReLU, além de uma camada de saída com 10 neurônios, correspondendo às classes do Fashion-MNIST. O treinamento ocorre por 10 épocas para cada método, registrando a perda e a acurácia. Ao final de cada época, o modelo é avaliado no conjunto de teste. Os resultados são apresentados por meio de gráficos de perda, que mostram a convergência do modelo, e gráficos de acurácia, que permitem comparar o desempenho dos três métodos ao longo do treinamento.

```python
import torch
import torch.nn as nn
import torch.optim as optim
import torchvision
import torchvision.transforms as transforms
import matplotlib.pyplot as plt
import numpy as np

# Definir semente para reprodutibilidade
SEED = 12345
torch.manual_seed(SEED)
np.random.seed(SEED)

# Hiperparâmetros
EPOCHS = 200  # Número de épocas
BATCH_SIZE = 64  # Tamanho do mini-batch
LEARNING_RATE = 0.01  # Taxa de aprendizado

# Transformação e normalização do dataset
transform = transforms.Compose([transforms.ToTensor(), transforms.Normalize((0.5,), (0.5,))])

# Carregar o dataset Fashion-MNIST
train_dataset = torchvision.datasets.FashionMNIST(root='./data', train=True, transform=transform, download=True)
test_dataset = torchvision.datasets.FashionMNIST(root='./data', train=False, transform=transform, download=True)

# Criar DataLoaders para cada método
train_loader_batch = torch.utils.data.DataLoader(dataset=train_dataset, batch_size=len(train_dataset), shuffle=True)
train_loader_sgd = torch.utils.data.DataLoader(dataset=train_dataset, batch_size=1, shuffle=True)
train_loader_minibatch = torch.utils.data.DataLoader(dataset=train_dataset, batch_size=BATCH_SIZE, shuffle=True)
test_loader = torch.utils.data.DataLoader(dataset=test_dataset, batch_size=1000, shuffle=False)

# Definir a estrutura da Rede Neural MLP
class MLP(nn.Module):
    def __init__(self):
        super(MLP, self).__init__()
        self.fc1 = nn.Linear(28*28, 128)  # Primeira camada totalmente conectada
        self.fc2 = nn.Linear(128, 64)  # Segunda camada
        self.fc3 = nn.Linear(64, 10)  # Camada de saída (10 classes)

    def forward(self, x):
        x = x.view(-1, 28*28)  # Flatten da imagem 28x28
        x = torch.relu(self.fc1(x))
        x = torch.relu(self.fc2(x))
        x = self.fc3(x)
        return x

# Função para treinar e avaliar o modelo
def train_and_evaluate(train_loader, method_name):
    """
    Treina um modelo MLP usando um dos métodos de gradiente descendente (Batch, SGD, Mini-Batch)
    e avalia seu desempenho.

    Parâmetros:
        train_loader: DataLoader correspondente ao método escolhido.
        method_name: Nome do método (para identificação nos gráficos).
    """
    model = MLP()  # Criar uma nova instância do modelo
    criterion = nn.CrossEntropyLoss()  # Função de perda
    optimizer = optim.SGD(model.parameters(), lr=LEARNING_RATE)  # Otimizador SGD padrão

    train_losses = []
    test_accuracies = []

    # Loop de treinamento
    for epoch in range(EPOCHS):
        model.train()
        epoch_loss = 0

        for images, labels in train_loader:
            optimizer.zero_grad()  # Zerar gradientes acumulados
            outputs = model(images)  # Forward
            loss = criterion(outputs, labels)  # Cálculo da perda
            loss.backward()  # Backpropagation
            optimizer.step()  # Atualização dos pesos

            epoch_loss += loss.item()

        # Média da perda por época
        train_losses.append(epoch_loss / len(train_loader))

        # Avaliação do modelo
        model.eval()
        correct = 0
        total = 0
        with torch.no_grad():
            for images, labels in test_loader:
                outputs = model(images)
                _, predicted = torch.max(outputs, 1)
                total += labels.size(0)
                correct += (predicted == labels).sum().item()

        accuracy = correct / total
        test_accuracies.append(accuracy)
        print(f'{method_name} - Época [{epoch+1}/{EPOCHS}], Loss: {train_losses[-1]:.4f}, Acurácia: {accuracy:.4f}')

    return train_losses, test_accuracies

# Executar os experimentos para cada método
print("\nTreinando com Gradiente Descendente em Batch:")
losses_batch, acc_batch = train_and_evaluate(train_loader_batch, "Batch")

print("\nTreinando com Gradiente Descendente Estocástico (SGD):")
losses_sgd, acc_sgd = train_and_evaluate(train_loader_sgd, "SGD")

print("\nTreinando com Gradiente Descendente Mini-Batch:")
losses_minibatch, acc_minibatch = train_and_evaluate(train_loader_minibatch, "Mini-Batch")

# Plotar os resultados
plt.figure(figsize=(12, 5))

# Gráfico de perda
plt.subplot(1, 2, 1)
plt.plot(losses_batch, label='Batch')
plt.plot(losses_sgd, label='SGD')
plt.plot(losses_minibatch, label='Mini-Batch')
plt.xlabel('Épocas')
plt.ylabel('Loss')
plt.title('Evolução da Perda')
plt.legend()

# Gráfico de acurácia
plt.subplot(1, 2, 2)
plt.plot(acc_batch, label='Batch')
plt.plot(acc_sgd, label='SGD')
plt.plot(acc_minibatch, label='Mini-Batch')
plt.xlabel('Épocas')
plt.ylabel('Acurácia')
plt.title('Evolução da Acurácia')
plt.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt

# Criar gráficos individuais para a função de custo de cada método
plt.figure(figsize=(8, 5))
plt.plot(losses_batch, label="Batch", color="blue")
plt.xlabel("Épocas")
plt.ylabel("Loss")
plt.title("Evolução da Perda - Batch")
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize=(8, 5))
plt.plot(losses_sgd, label="SGD", color="red")
plt.xlabel("Épocas")
plt.ylabel("Loss")
plt.title("Evolução da Perda - SGD")
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize=(8, 5))
plt.plot(losses_minibatch, label="Mini-Batch", color="green")
plt.xlabel("Épocas")
plt.ylabel("Loss")
plt.title("Evolução da Perda - Mini-Batch")
plt.legend()
plt.grid(True)
plt.show()

# Comparação dos Resultados - Gráficos de perda e acurácia
plt.figure(figsize=(12, 5))

# Curvas de perda para treino
plt.subplot(1, 2, 1)
plt.plot(losses_batch, label="Batch", color="blue")
plt.plot(losses_sgd, label="SGD", color="red")
plt.plot(losses_minibatch, label="Mini-Batch", color="green")
plt.xlabel("Épocas")
plt.ylabel("Loss")
plt.title("Curvas de Perda para Treinamento")
plt.legend()
plt.grid(True)

# Curvas de acurácia para validação
plt.subplot(1, 2, 2)
plt.plot(acc_batch, label="Batch", color="blue")
plt.plot(acc_sgd, label="SGD", color="red")
plt.plot(acc_minibatch, label="Mini-Batch", color="green")
plt.xlabel("Épocas")
plt.ylabel("Acurácia")
plt.title("Curvas de Acurácia para Validação")
plt.legend()
plt.grid(True)

plt.show()

# Avaliação final dos modelos usando métricas padrão (Acurácia final e Erro médio final)
import pandas as pd
import seaborn as sns

# Criar DataFrame para comparação das métricas finais
metrics_df = pd.DataFrame({
    "Método": ["Batch", "SGD", "Mini-Batch"],
    "Acurácia Final": [acc_batch[-1], acc_sgd[-1], acc_minibatch[-1]],
    "Erro Médio Final": [losses_batch[-1], losses_sgd[-1], losses_minibatch[-1]]
})

# Gráfico de barras para Acurácia final
plt.figure(figsize=(8, 5))
sns.barplot(x="Método", y="Acurácia Final", data=metrics_df, palette="viridis")
plt.title("Comparação da Acurácia Final dos Modelos")
plt.ylabel("Acurácia")
plt.xlabel("Método")
plt.ylim(0, 1)
plt.grid(True)
plt.show()

# Gráfico de barras para Erro médio final
plt.figure(figsize=(8, 5))
sns.barplot(x="Método", y="Erro Médio Final", data=metrics_df, palette="magma")
plt.title("Comparação do Erro Médio Final dos Modelos")
plt.ylabel("Erro Médio")
plt.xlabel("Método")
plt.grid(True)
plt.show()


```

A comparação entre os métodos de Gradiente Descendente Batch, Mini-Batch e Estocástico (SGD) permite analisar como cada abordagem influencia a convergência, estabilidade e desempenho do modelo no dataset Fashion-MNIST.

##1. Convergência e Estabilidade

O Gradiente Descendente em Batch apresentou uma curva de perda suave e estável, refletindo a característica esperada desse método. Como a atualização dos pesos ocorre após o cálculo do gradiente sobre todo o conjunto de treinamento, ele segue um caminho mais estável para o mínimo da função de custo. No entanto, a curva de perda do Batch demonstra uma descida muito lenta, o que sugere que o modelo está aprendendo de forma ineficiente, possivelmente devido à baixa frequência de atualização dos pesos e à necessidade de um learning rate maior. Além disso, a curva de acurácia do Batch indica que o modelo está aprendendo devagar, alcançando um desempenho significativamente inferior aos demais métodos dentro do mesmo número de épocas.

Já o Gradiente Descendente Estocástico (SGD) mostrou uma convergência mais rápida, porém com oscilações notáveis. Como as atualizações ocorrem após cada amostra individualmente, há maior variabilidade na direção do gradiente, o que pode ajudar o modelo a escapar de mínimos locais ruins, mas também pode dificultar a estabilização do aprendizado. Essa característica pode ser observada na curva de perda, que apresenta flutuações mais significativas em relação aos outros métodos.

O Gradiente Descendente Mini-Batch mostrou-se um compromisso ideal entre os dois extremos. Ele equilibrou a estabilidade do Batch com a eficiência computacional do SGD, realizando atualizações mais frequentes do que o Batch, mas mantendo uma suavidade relativa na convergência. Isso permitiu um aprendizado rápido e estável, o que é refletido na curva de acurácia, que alcança valores próximos a 90% rapidamente e mantém um desempenho estável ao longo do treinamento. Esse comportamento confirma o motivo pelo qual o Mini-Batch é amplamente utilizado na prática para treinamento de redes neurais profundas.

##2. Precisão Final e Eficiência
A acurácia final dos três métodos apresentou algumas diferenças notáveis:

- O Batch teve uma convergência muito lenta, sugerindo que o modelo pode ter atingido um platô antes de explorar melhor o espaço de otimização.
- O SGD, devido à sua aleatoriedade, apresentou uma convergência rápida, mas com maior instabilidade ao longo do tempo.
- O Mini-Batch mostrou-se o mais eficiente, atingindo alta acurácia rapidamente sem oscilações significativas, tornando-se uma abordagem equilibrada.

Além disso, a eficiência computacional também deve ser considerada. O método Batch, apesar de estável, exige maior uso de memória, pois processa todo o dataset antes de atualizar os pesos. O SGD, ao contrário, atualiza os pesos constantemente e exige menos memória, mas pode ser computacionalmente caro devido ao grande número de atualizações. O Mini-Batch equilibra essas questões, oferecendo bons resultados de convergência com menor custo computacional, o que o torna uma abordagem amplamente utilizada na prática.

Dessa forma, dado o comportamento observado nos gráficos, o Batch não é recomendado nesse cenário, pois apresenta desempenho inferior e convergência muito lenta. O SGD pode ser útil para escapar de mínimos locais, mas sofre com instabilidade. O Mini-Batch mostrou ser a opção mais eficiente, combinando aprendizado rápido, estabilidade e eficiência computacional.

##3. Impacto da Escolha do Método na Convergência e Estabilidade
A escolha do método de gradiente descendente tem um impacto direto na eficiência do treinamento e na capacidade do modelo de generalizar bem para novos dados. O Batch pode ser uma opção viável quando há alto poder computacional e tempo suficiente para um aprendizado mais lento e estável, mas, no caso analisado, mostrou-se ineficiente. O SGD, apesar de sua instabilidade, pode ser vantajoso para problemas onde escapar de mínimos locais é essencial. No entanto, o Mini-Batch se destacou como a abordagem mais eficaz, pois combinou a eficiência do SGD com a estabilidade do Batch, atingindo bons resultados em menos tempo, tornando-se a melhor escolha para treinar redes neurais profundas.



#8.Conclusão
Os experimentos demonstraram que cada método de Gradiente Descendente apresenta vantagens e desvantagens distintas, influenciando diretamente a convergência, estabilidade e eficiência computacional do modelo. Entre as abordagens analisadas, o Mini-Batch mostrou-se a melhor opção, pois atingiu alta acurácia rapidamente, com menos oscilações do que o SGD e treinamento significativamente mais eficiente do que o Batch.

O Batch Gradient Descent, apesar de apresentar uma curva de perda estável, teve um aprendizado muito lento, indicando que sua aplicação pode ser limitada em cenários onde é necessário um ajuste mais dinâmico dos pesos. Já o SGD, por sua vez, proporcionou um aprendizado rápido, mas com maior instabilidade, o que pode ser um problema em tarefas que exigem alta precisão.

Portanto, o Mini-Batch Gradient Descent é a alternativa mais equilibrada, pois combina boa estabilidade, rapidez na convergência e menor custo computacional, tornando-se a escolha mais prática e eficiente para treinamento de redes neurais profundas. No entanto, dependendo do contexto do problema e dos recursos disponíveis, o Batch ou o SGD ainda podem ser opções viáveis, desde que suas características sejam bem compreendidas e ajustadas conforme necessário.
